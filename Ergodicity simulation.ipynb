```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
sims = np.random.randint(2,size=(50,1000))

win = 1.45
lose = 0.6

sims = np.where(sims == 1, win, lose)
sims[0]
```

```python
ensemble = np.cumproduct(sims,axis=0)
ensemble
```

```python
plt.plot(ensemble)
plt.yscale('log')
plt.show()
```

```python
final = ensemble[-1,:]
print(np.median(final),np.mean(final))
```

```python
plt.hist(final, 100, density=True)
plt.xscale('log')
plt.show()
```
